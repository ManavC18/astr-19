```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
random_numbers = np.random.noncentral_chisquare(df=3, nonc=5, size=1000)
```

```python
plt.hist(random_numbers, bins=100, color='mediumpurple', edgecolor='black', alpha=0.8)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('1000 Random Variates: Noncentral Chi-Square (df=3, nonc=5)')

plt.savefig("session_10_plot.pdf", bbox_inches='tight')

plt.show()
```
