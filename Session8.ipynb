```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def get_sin(x):
    return np.sin(x)

def get_cos(x):
    return np.cos(x)
```

```python
f, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(x, get_sin(x))
ax[0].set_xlabel('x')
ax[0].set_ylabel('sin(x)')
ax[0].set_title('sin(x)')

ax[1].plot(x, get_cos(x))
ax[1].set_xlabel('x')
ax[1].set_ylabel('cos(x)')
ax[1].set_title('cos(x)')

plt.tight_layout()

plt.savefig("session_8_plot.pdf", bbox_inches='tight')

plt.show()
```
