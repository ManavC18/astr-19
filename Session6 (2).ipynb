### Import Statements
Importing the required modules for this notebook

```python
import numpy as np
import math
from tabulate import tabulate    
```

### Sine Function
This cell executes a sine operation on the variable x

```python
def sinfunc(x):
    return math.sin(x)
```

### Cosine Function
This cell executes a cosine operation on the variable x

```python
def cosfunc(x):
    return math.cos(x)
```

### Tabulate Data
This cell loops a variable x 1000 times over an interval from 0 to 2pi, uses the sin and cos functions to add the sin and cos values of each x value to their respective lists to organize the data, and finally calls the tabulate module to organize the data in a table

```python
xvals = np.linspace(0.0, 2 * math.pi, 1000)

datatable = []
for x in xvals:
    datatable.append((x, sinfunc(x), cosfunc(x)))

headers = ["x", "sin(x)", "cos(x)"]
print(tabulate(datatable, headers = headers, tablefmt = "grid", floatfmt = ".3f"))
```

### Prints First 10 Entries
This cell prints the first 10 entries of x, sin(x), and cos(x) in columns

```python
for i in range(10):
    print(f"{xvals[i]:.3f}\t{sinfunc(xvals[i]):.3f}\t{cosfunc(xvals[i]):.3f}")
```
